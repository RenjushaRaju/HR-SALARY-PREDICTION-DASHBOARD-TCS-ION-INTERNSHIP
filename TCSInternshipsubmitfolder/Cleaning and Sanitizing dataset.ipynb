```python
from datetime import date
def experienceCalc(doj, dot):
    today = date.today()
    return np.where(pd.isna(dot), today.year - doj.dt.year, dot.dt.year - doj.dt.year)
 
df['YearsOfExperience'] = experienceCalc(df['DateofHire'], df['DateofTermination'])
df[['YearsOfExperience']]
```

```python
df['Date'] = pd.to_datetime(df.DOB)

df['Date']
```

```python
from datetime import date
def from_dob_to_age(born):
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
```

```python
df['Age']=df['Date'].apply(lambda x: from_dob_to_age(x))
```

```python
#outlier removal
import numpy as np

Q1=np.percentile(df['Salary'],25,interpolation='midpoint')
Q2=np.percentile(df['Salary'],50,interpolation='midpoint')
Q3=np.percentile(df['Salary'],75,interpolation='midpoint')
print(Q1)
print(Q2)
print(Q3)
df['Salary'].median()
IQR=Q3-Q1
low_lim=Q1-1.5*IQR
Up_lim=Q3+1.5*IQR
print(low_lim)
print(Up_lim)
outlier=[]
```

```python

for x in df['Salary']:
    if((x>Up_lim)or(x<low_lim)):
        outlier.append(x)
outlier        


ind1=df['Salary']>Up_lim
in2=df.loc[ind1].index
df.drop(in2,inplace=True)
ind2=df['Salary']<low_lim
in3=df.loc[ind2].index
df.drop(in3,inplace=True)



plt.boxplot(df['Salary'])
plt.title('Boxplot of Salary')
```
